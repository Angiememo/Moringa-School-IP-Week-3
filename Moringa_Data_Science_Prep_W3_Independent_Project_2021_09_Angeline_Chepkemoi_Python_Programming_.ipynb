# Importing our pandas library

```python
# Let's import the pandas library just like we did with Numpy
# 
import pandas as pd
```

```python
# We are going to read the data from the links given.
# From cells_geo.csv
df_cells = pd.read_csv('cells_geo.csv', delimiter=';', header= 1, names= ['Cities', 'Status', 'Localisation', 'Decoupzone','Zonename', 'Longitude', 'Latitude', 'Region', 'Area', 'Cell_ID', 'Site_ID'])
df_cells
```

```python
# From Telkom Dataset1
df_1 = pd.read_csv('Telcom_dataset.csv', header = 1, names = ['Product','Value','Datetime','Cell_on_Site','Dw_A_Number_Int','Dw_B_Number_Int','Country_A','Country_B','Cell_ID','Site_ID'])
df_1
```

```python
# From Telkom Dataset2
df_2 = pd.read_csv('Telcom_dataset2.csv', header = 1, names = ['Product','Value','Datetime','Cell_on_Site','Dw_A_Number_Int','Dw_B_Number_Int','Country_A','Country_B','Cell_ID','Site_ID'])
df_2
```

```python
# From Telkom Dataset3
df_3 = pd.read_csv('Telcom_dataset3.csv', header = 1, names = ['Product','Value','Datetime','Cell_on_Site','Dw_A_Number_Int','Dw_B_Number_Int','Country_A','Country_B','Cell_ID','Site_ID'])
df_3
```

# Merging the Datasets

```python
# We are going to merge the telkom datasets

merged_datasets = pd.concat([df_1, df_2, df_3],axis=0)
merged_datasets
```

```python
# Changing the 'datetime' column to 'datetime' format

merged_datasets['Datetime'] = pd.to_datetime(merged_datasets['Datetime'])
merged_datasets
```

```python
# Merging the four cells

merged_datasets_2=merged_datasets.merge(df_cells, how='left', on='Cell_ID')
merged_datasets_2
```

# Finding the most used city

```python
# The most used city for the three days.

city_use = merged_datasets_2['Value'].groupby(merged_datasets_2['Cell_ID'].rank(ascending=True))
city_use.mean()
```

```python
# The most used product.

product_use = merged_datasets_2['Value'].groupby(merged_datasets_2['Product'])
product_use.mean()
```
